```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.metrics import r2_score , mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
X = pd.DataFrame(data["Hours"])
y = pd.DataFrame(data["Scores"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

# I have used 3 algorithms to predict
## One without regularization and two with Regularization

```python
lm1 = LinearRegression()
lm2 = Lasso()
lm3 = Ridge()
```

```python
lm1.fit(X_train,y_train)
lm2.fit(X_train,y_train)
lm3.fit(X_train,y_train)
```

```python
pred1 =lm1.predict(X_test)
pred2 =lm2.predict(X_test)
pred3 =lm3.predict(X_test)
```

```python
r2_score(pred1,y_test)
```

```python
r2_score(pred2,y_test)
```

```python
r2_score(pred3,y_test)
```
